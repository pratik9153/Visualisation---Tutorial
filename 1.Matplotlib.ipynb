```python
#import matplotlib
import matplotlib.pyplot as plt
```

```python
#import numpy
import numpy as np
```

```python
#generate  100 data between 0 to 100 using linspace
x = np.linspace(0,10 , (200))
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python

plt.figure(figsize = (6,2))
plt.plot(x,y,'--b')
plt.xlabel("this is my x data") #Label for X axis
plt.ylabel("this is my y data") #label for y axis
plt.title("this is my title") #title 
plt.grid()
plt.show()
```

```python
# Generating Random value using numpy and scatter the data point
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50) # colours for data point
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c = colours , s = sizes , alpha = .8) #Alpha in Matplotlib : Adjusts Transparency.
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("random data")
```

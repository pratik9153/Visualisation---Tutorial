```python
#import seaborn as sns 
import seaborn as sns
```

```python
#load dataset iris from seaborn
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# scatterplot 

sns.scatterplot(x=df.sepal_length , y=df.sepal_width)
```

```python
sns.scatterplot(x=df.sepal_length , y=df.petal_length)
```

```python
sns.scatterplot(x=df.petal_width , y=df.petal_length)
```

```python
#using displot to observ data distribution 
sns.displot(df['sepal_length'])
```

```python
#load data set tip
df1 = sns.load_dataset('tips')
```

```python
df1.head()
```

```python
sns.scatterplot(x=df1.total_bill , y = df1.tip)
```

```python
# to display categories in smoker columns
df1['smoker'].value_counts()
```

```python
#Redraw plot with new data.
sns.relplot(x = df1.total_bill , y = df1.tip , data = df1 , hue = 'smoker')
```

```python
sns.relplot(x = df1.total_bill , y = df1.tip , data = df1 , style = 'size',hue = 'time')
```

```python
sns.relplot(x = df1.total_bill , y = df1.tip , data = df1 , hue = 'time')
```

```python
df1['size'].value_counts()
```

```python
#categorical plot
sns.catplot(x='day' , y = 'total_bill' ,data = df1)
```

```python
sns.catplot(x='smoker' , y = 'tip' ,data = df1)
```

```python
#Joint distribution plot 
sns.jointplot(x=df1.total_bill , y = df1.tip)
```

```python
#Plot pairwise relationships in dataset.
sns.pairplot(df1)
```
